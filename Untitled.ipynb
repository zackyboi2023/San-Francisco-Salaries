```python
import kaggle
import zipfile
import sqlite3

import pandas as pd
import numpy as np
import warnings

import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
zipfile_data = 'sf-salaries.zip'
with zipfile.ZipFile(zipfile_data, 'r') as file:
    file.extractall()
```

```python
#Connecting to the the database

con = sqlite3.connect(r"database.sqlite")
cursor = con.cursor()
```

```python
#read the SQL file

data = pd.read_sql("""SELECT *
                      FROM sqlite_master;
                        """, con)

data
```

```python
#get the important database table

salariesData = pd.read_sql("""SELECT *
                              FROM Salaries;
                                """, con)

salariesData
```

## Data Cleaning

### Cleaning Nan values

```python
salariesData[salariesData['Benefits'] == 'Not Provided'].index
```

```python
salariesData.drop([148646, 148650, 148651, 148652], inplace=True)
```

```python
salariesData['BasePay'] = pd.to_numeric(salariesData['BasePay'])
salariesData['OvertimePay'] = pd.to_numeric(salariesData['OvertimePay'])
salariesData['OtherPay'] = pd.to_numeric(salariesData['OtherPay'])
salariesData['Benefits'] = pd.to_numeric(salariesData['Benefits'])
```

```python
for col in salariesData.columns:
    salariesData[col] = salariesData[col].replace("", "No Value")
```

```python
salariesData
```

```python
warnings.filterwarnings('ignore', category = FutureWarning)

# Iterate over columns
for col in salariesData.columns:
    # Check if "No Value" is in any element of the current column
    if "No Value" in salariesData[col].values:
        salariesData.drop(col, axis = 1, inplace = True)
```

```python
salariesData
```

```python
salariesData.isna().sum()
```

```python
columns =['BasePay','OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']

for col in columns:
    if salariesData[col].isna().any():
        salariesData[col].fillna(0, inplace = True)
```

```python
salariesData.isna().sum()
```

### Cleaning Strings

```python
salariesData['EmployeeName'] = salariesData['EmployeeName'].str.title()
```

```python
salariesData['JobTitle'] = salariesData['JobTitle'].str.title()
```

```python
salariesData
```

### Are there any duplicates?

```python
salariesData.duplicated(keep = False, subset = ['EmployeeName', 'JobTitle'])
```

```python
salariesData.drop_duplicates(keep = "first", 
                             subset = ['EmployeeName', 'JobTitle'], 
                             inplace = True)
```

### What is the distribution of Base pay to total benefits within san fransico?

```python
payCols = ['BasePay', 'Benefits', 'TotalPay', 'TotalPayBenefits']

for col in payCols:
    plt.figure(figsize = (14, 7))
    
    sbn.histplot(data = salariesData, x = col, kde = True)
    
    plt.title(f"The Distribution of {col}")
    plt.show()
```

### How many jobs where created in 2014

### Create a scatter graph based on the pay to benefits

```python
plt.figure(figsize = (14, 7))

sbn.scatterplot(data = salariesData, x = 'BasePay', y = 'Benefits')

# Calculate the average values
avg_base_pay = salariesData['BasePay'].mean()
avg_benefits = salariesData['Benefits'].mean()

# Plot horizontal and vertical lines at the average points
plt.axvline(x=avg_base_pay, color='red', label=f'Average BasePay: {avg_base_pay:.2f}')
plt.axhline(y=avg_benefits, color='blue', label=f'Average Benefits: {avg_benefits:.2f}')

plt.title("Scatter graph based on the pay to benefits")
plt.show()
```
